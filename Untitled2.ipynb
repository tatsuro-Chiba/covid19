```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# 空気抵抗を考慮した弾道計算関数
def calculate_trajectory(v0, angle, size, density, Cd, target_distance, target_height, g=9.81, rho_air=1.225):

    v = v0 * np.sin(np.radians(angle)) ** 1.5
    angle_rad = np.radians(angle)
    vx = v * np.cos(angle_rad)
    vy = v * np.sin(angle_rad)
    initial_conditions = [0, 0, vx, vy]

    def equations_of_motion(t, y):
        x, y, vx, vy = y
        v = np.sqrt(vx**2 + vy**2)
        Fd = 0.5 * Cd * rho_air * np.pi * (size/2)**2 * v**2
        dvxdt = -Fd * vx / v / (density * (4/3) * np.pi * (size/2)**3)
        dvydt = -g - Fd * vy / v / (density * (4/3) * np.pi * (size/2)**3)
        return [vx, vy, dvxdt, dvydt]

    # ODEソルバーを使用して軌跡を計算
    result = solve_ivp(equations_of_motion, [0, target_distance/vx*2], initial_conditions, t_eval=np.linspace(0, target_distance/vx*2, 1000))
    x, y, vx, vy = result.y

    # ターゲット地点に到達するかチェック
    for i in range(len(x)):
        if x[i] >= target_distance and y[i] >= target_height:
            final_vx = vx[i]
            final_vy = vy[i]
            final_speed = np.sqrt(final_vx**2 + final_vy**2)
            final_angle = np.degrees(np.arctan2(final_vy, final_vx))
            return True, x[:i+1], y[:i+1], final_speed, final_angle
    return False, x, y, None, None

####################################################

# 弾道計算のパラメータ 球体
size = 0.3    # 弾のサイズ (m)
density = 2200  # 弾の密度 (kg/m^3)
Cd = 0.47      # 抗力係数

mode=2

if mode==1:  #御嶽西
  #火口の座標 m
  vent_x = -92444
  vent_y = -11609
  vent_z = 2770
  #目標地点の座標 m
  target_x = -92556
  target_y = -11482
  target_z = 2924
  size=0.95

if mode==2:  #御嶽東
  #火口の座標  m
  vent_x = -92444
  vent_y = -11609
  vent_z = 2770
  #目標地点の座標 m
  target_x = -91797
  target_y = -11237
  target_z = 2988
  size=0.5
#############

# 2点間の水平距離を計算
target_distance = np.sqrt((target_x - vent_x)**2 + (target_y - vent_y)**2)
# 高度差を計算
target_height = target_z - vent_z

# 目標地点の設定 ここを書き換えてもよい
#target_distance = 500  # 距離 (m)　火口から落下点までの水平距離
#target_height = 0     # 比高 (m)　プラスが投げ上げ

####################################################

v0 = 50  # 初速 (m/s)　探索計算開始速度

while True:
    for angle in range(40, 71):
        success, x, y, final_speed, final_angle = calculate_trajectory(v0, angle, size, density, Cd, target_distance, target_height)
        if success:
            # ラベルのフォーマットを更新
            v = v0 * np.sin(np.radians(angle)) ** 1.5
            label_text = (
                f"Particle Size (Diameter): {size} m\n"
                f"Particle Density: {density} kg/m³\n"
                f"Cd Value: {Cd}\n"
                f"Vent-target horizontal distance: {target_distance} m\n"
                f"Elevation difference of vent and target: {target_height} m\n\n"
                f"Initial Vartical max Velocity: {v0:.2f} m/s\n"
                f"Injection Angle: {angle} degrees\n"
                f"Injection Velocity: {v:.2f} m/s\n"
                f"Final impact Angle: {final_angle:.2f} degrees\n"
                f"Final impact Velocity: {final_speed:.2f} m/s\n"
            )
            # 軌跡をプロット
            plt.figure(figsize=(10, 6))
            plt.plot(x, y, label=label_text)
            plt.scatter([target_distance], [target_height], color='red')  # 目標地点
            plt.gca().set_aspect('equal', adjustable='box')
            plt.title('Trajectory with Air Resistance')
            plt.xlabel('Distance (m)')
            plt.ylabel('Height (m)')
            plt.ylim(bottom=-200)
            #plt.xlim(right=400)
            plt.grid(True)
            plt.legend()
            plt.show()
            break
    else:
        v0 += 0.5  # 初速を増加
        continue
    break
```
