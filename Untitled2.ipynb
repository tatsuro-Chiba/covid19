```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt


######################　2024-1-6　　#これまでの結果と会わないなあ　mode=5


# 組み込み済みのコードに空気密度計算機能を追加する

def calculate_trajectory(v0, angle, size, density, Cd, target_distance, target_height, vent_z, T_sea_level, P_sea_level, g=9.81):
    R = 287.05  # Specific gas constant for dry air, J/(kg·K)
    L = 0.0065  # Standard temperature lapse rate, K/m

    # Convert sea level temperature to Kelvin
    T_sea_level_K = T_sea_level + 273.15

    # Calculate temperature at vent altitude
    T_vent = T_sea_level_K - L * vent_z

    # Calculate pressure at vent altitude
    P_vent = P_sea_level * (1 - L * vent_z / T_sea_level_K) ** (g / (R * L))

    # Calculate air density at vent altitude
    rho_air = P_vent / (R * T_vent)

    # Adjusted initial velocity considering the angle
    v = v0 * np.sin(np.radians(angle)) ** 1.5
    angle_rad = np.radians(angle)
    vx = v * np.cos(angle_rad)
    vy = v * np.sin(angle_rad)
    initial_conditions = [0, 0, vx, vy]

    def equations_of_motion(t, y):
        x, y, vx, vy = y
        v = np.sqrt(vx**2 + vy**2)
        Fd = 0.5 * Cd * rho_air * np.pi * (size/2)**2 * v**2
        dvxdt = -Fd * vx / v / (density * (4/3) * np.pi * (size/2)**3)
        dvydt = -g - Fd * vy / v / (density * (4/3) * np.pi * (size/2)**3)
        return [vx, vy, dvxdt, dvydt]

    # ODE solver to calculate trajectory
    result = solve_ivp(equations_of_motion, [0, target_distance/vx*2], initial_conditions, t_eval=np.linspace(0, target_distance/vx*2, 1000))
    x, y, vx, vy = result.y

    # Check if the target point is reached
    for i in range(len(x)):
        if x[i] >= target_distance and y[i] >= target_height:
            final_vx = vx[i]
            final_vy = vy[i]
            final_speed = np.sqrt(final_vx**2 + final_vy**2)
            final_angle = np.degrees(np.arctan2(final_vy, final_vx))
            return True, x[:i+1], y[:i+1], final_speed, final_angle
    return False, x, y, None, None


####################################################

# 弾道計算のパラメータ 球体

density = 2200  # 弾の密度 (kg/m^3)
Cd = 0.47      # 抗力係数
T_sea_level=12
P_sea_level=1020

mode=4

if mode==1: #御嶽山西
  vent_x = -92444
  vent_y = -11609
  vent_z = 2770
  target_x = -92556
  target_y = -11482
  target_z = 2924
  size = 0.94    # 弾のサイズ (m)

if mode==2: #御嶽山東
  vent_x = -92444
  vent_y = -11609
  vent_z = 2770
  target_x = -91797
  target_y = -11237
  target_z = 2988
  size = 0.5    # 弾のサイズ (m)

if mode==3: #草津西
  vent_x = -115687
  vent_y = 70449
  vent_z = 2035
  target_x = -115400
  target_y = 70717
  target_z = 1928
  size = 1.0    # 弾のサイズ (m)

if mode==4: #草津東
  vent_x = -115687
  vent_y = 70449
  vent_z = 2035
  target_x = -115373
  target_y = 70361
  target_z = 1979
  size = 0.5    # 弾のサイズ (m)

if mode==5: #草津北東
  vent_x = -115687
  vent_y = 70449
  vent_z = 2035
  target_x = -115088
  target_y = 70728
  target_z = 1841
  size = 0.2   # 弾のサイズ (m)

# 2点間の水平距離を計算
target_distance = np.sqrt((target_x - vent_x)**2 + (target_y - vent_y)**2)
# 高度差を計算
target_height = target_z - vent_z

# 目標地点の設定 ここを書き換えてもよい
#target_distance = 500  # 距離 (m)　火口から落下点までの水平距離
#target_height = 0     # 比高 (m)　プラスが投げ上げ

####################################################

v0 = 50  # 初速 (m/s)　探索計算開始速度

while True:
    for angle in range(40, 71):
        #success, x, y, final_speed, final_angle = calculate_trajectory(v0, angle, size, density, Cd, target_distance, target_height)

        success, x, y, final_speed, final_angle = calculate_trajectory(
            v0, angle, size, density, Cd, target_distance, target_height,
            vent_z, T_sea_level, P_sea_level
        )



        if success:
            # ラベルのフォーマットを更新
            v = v0 * np.sin(np.radians(angle)) ** 1.5
            label_text = (
                f"   Air pressure (sea level): {P_sea_level} pascal\n"
                f"   Temparature (sea level) : {T_sea_level} centi degree\n"
                f"Particle Size (Diameter): {size} m\n"
                f"Particle Density: {density} kg/m3\n"
                f"Cd Value: {Cd}\n"
                f"Vent-target horizontal distance: {target_distance:.0f} m\n"
                f"Elevation difference of vent and target: {target_height} m\n\n"
                f"Initial vartical max velocity: {v0:.2f} m/s\n"
                f"Injection Angle: {angle} degrees\n"
                f"Injection velocity: {v:.2f} m/s\n"
                f"Final impact Angle: {final_angle:.2f} degrees\n"
                f"Final impact velocity: {final_speed:.2f} m/s\n"
            )
            # 軌跡をプロット
            plt.figure(figsize=(10, 6))
            plt.plot(x, y, label=label_text)
            plt.scatter([target_distance], [target_height], color='red')  # 目標地点
            plt.gca().set_aspect('equal', adjustable='box')
            plt.title('Trajectory with Air Resistance')
            plt.xlabel('Distance (m)')
            plt.ylabel('Height (m)')
            plt.ylim(bottom=-250)
            plt.grid(True)
            plt.legend()
            plt.show()
            break
    else:
        v0 += 0.5  # 初速を増加
        continue
    break
```
