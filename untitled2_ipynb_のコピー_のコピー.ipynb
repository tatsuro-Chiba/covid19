```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.ndimage import map_coordinates
from osgeo import gdal
from matplotlib.colors import LightSource, LinearSegmentedColormap
from google.colab import drive
from scipy.interpolate import interp1d
from scipy.integrate import solve_ivp
import matplotlib.cm as cm
from datetime import datetime

#############################################################################

drive.mount('/content/drive')

grey_value = 150 / 255.0
custom_cmap = LinearSegmentedColormap.from_list('custom_grey', [(grey_value, grey_value, grey_value), (grey_value, grey_value, grey_value)])


def adjust_azimuth(azimuth_deg):
    """
    北を0度とし、東を90度とするための方位角の調整
    """
    az = 90-azimuth_deg
    if az < 0:
        az += 360
    return az

# 地形データの読み込み
def load_dem_data(dem_file_path):
    dem_data = gdal.Open(dem_file_path)
    geo_transform = dem_data.GetGeoTransform()
    dem_array = dem_data.ReadAsArray()
    return dem_array, geo_transform

# 地形断面の計算
def calculate_terrain_profile(vent_x, vent_y, azimuth_deg, distance, dem_array, geo_transform, sampling_interval=1):
    azimuth_rad = np.deg2rad(azimuth_deg)
    distances = np.arange(0, distance + 1, sampling_interval)
    x_rel = vent_x + np.cos(azimuth_rad) * distances
    y_rel = vent_y + np.sin(azimuth_rad) * distances
    px, py = (x_rel - geo_transform[0]) / geo_transform[1], (y_rel - geo_transform[3]) / geo_transform[5]
    elevations = map_coordinates(dem_array, [py, px], order=1, mode='nearest')
    return distances, elevations

# 衝突点の検出
def detect_collision_point(df_trajectory, distances, elevations, vent_z, ignore_within=10):
    for index, row in df_trajectory.iterrows():
        x_pos = row['X Position (m)']
        z_pos = row['Z Position (m)'] + vent_z  # 火口の高度を考慮
        if x_pos <= ignore_within:
            continue
        nearest_idx = np.argmin(np.abs(distances - x_pos))
        terrain_z = elevations[nearest_idx]
        if z_pos <= terrain_z:
            return x_pos, z_pos
    return None, None

# 衝突点の情報を格納するためのDataFrameの作成
def init_collision_details_df():
    return pd.DataFrame(columns=['Azimuth', 'Angle', 'Geo X', 'Geo Y', 'X Position', 'Z Position', 'Velocity X', 'Velocity Z', 'Fall Speed', 'Fall Angle'])

# 全ての軌跡と衝突点を処理してプロットする
def plot_all_trajectories_and_terrain(df_results, dem_array, geo_transform, vent_x, vent_y, vent_z, azimuths):
    collision_details = init_collision_details_df()
    cmap = plt.get_cmap('hsv', len(df_results['Angle (deg)'].unique()))

    for azimuth in azimuths:
        fig, ax = plt.subplots(figsize=(10, 8))
        distances, elevations = calculate_terrain_profile(vent_x, vent_y, azimuth, df_results['X Position (m)'].max(), dem_array, geo_transform)
        ax.plot(distances, elevations, label='Terrain Profile', color='grey', linewidth=2)

        for i, angle in enumerate(sorted(df_results['Angle (deg)'].unique())):
            df_angle = df_results[df_results['Angle (deg)'] == angle]
            collision_x, collision_z = detect_collision_point(df_angle, distances, elevations, vent_z, ignore_within=10)

            if collision_x is not None and collision_z is not None:
                geo_x = vent_x + collision_x * np.cos(np.deg2rad(azimuth))
                geo_y = vent_y + collision_x * np.sin(np.deg2rad(azimuth))
                df_angle_before_collision = df_angle[df_angle['X Position (m)'] <= collision_x]
                color = cmap(i)
                ax.plot(df_angle_before_collision['X Position (m)'], df_angle_before_collision['Z Position (m)'] + vent_z, color=color, label=f'Angle: {angle}°')
                ax.scatter(collision_x, collision_z, color='red', zorder=5, marker='^', s=10)

                new_row = pd.DataFrame([{
                    'Azimuth': azimuth, 'Angle': angle,
                    'Geo X': geo_x, 'Geo Y': geo_y,
                    'X Position': collision_x, 'Z Position': collision_z,
                    'Velocity X': df_angle_before_collision.iloc[-1]['X Velocity (m/s)'],
                    'Velocity Z': df_angle_before_collision.iloc[-1]['Z Velocity (m/s)'],
                    'Fall Speed': df_angle_before_collision.iloc[-1]['Speed (m/s)'],
                    'Fall Angle': df_angle_before_collision.iloc[-1]['Flight Angle (deg)']
                }])
                collision_details = pd.concat([collision_details, new_row], ignore_index=True)
            else:
                ax.plot(df_angle['X Position (m)'], df_angle['Z Position (m)'] + vent_z, color=cmap(i), label=f'Angle: {angle}°')

        ax.axhline(y=vent_z, color='black', linestyle='--', label='Vent Elevation')
        ax.set_xlabel('Distance from Vent (m)')
        ax.set_ylabel('Elevation (m)')
        ax.set_title(f'Trajectry  (Azimuth {adjust_azimuth(azimuth)}°)')
        ax.grid(True)
        ax.axis('equal')
        ax.legend(loc='upper right', fontsize='small')
        plt.show()

    return collision_details

# 地形データの読み込みと陰影図の作成
def plot_shaded_dem(dem_array, geo_transform, points, color_by, title, cmap='viridis'):
    nrows, ncols = dem_array.shape
    x_min = geo_transform[0]
    y_max = geo_transform[3]
    x_max = x_min + geo_transform[1] * ncols
    y_min = y_max + geo_transform[5] * nrows

    ls = LightSource(azdeg=315, altdeg=45)
    shaded = ls.shade(dem_array, cmap=custom_cmap, blend_mode='overlay')

    plt.figure(figsize=(10, 8))
    plt.imshow(shaded, extent=(x_min, x_max, y_min, y_max))
    plt.scatter(points['Geo X'], points['Geo Y'], c=points[color_by], cmap=cmap, s=5)
    plt.colorbar(label=color_by)
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(title)
    plt.show()


# メイン処理

mode=1

if mode==1:
    dem_file_path = '/content/drive/My Drive/hakone/ontake_dem_file.tif'
    dem_array, geo_transform = load_dem_data(dem_file_path)
    csv_file = '/content/drive/My Drive/trajectory_simulation_results_20240311_102424.csv'  #V0dmin=207
    df_results = pd.read_csv(csv_file)
    vent_locations_list = [[1, -92444, -11609, 2770]]

if mode==2:
    dem_file_url = 'https://paspro.com/bousai/kazan/_/bbs/?file_id=MV8wMDA3NDI%3D&name=9kei-kusatsu-motoshirane-490cm-2.tif'
    dem_file_path = '/content/drive/My Drive/hakone/kusatsu_dem_file-3.tif'
    r = requests.get(dem_file_url)
    with open(dem_file_path, 'wb') as f:
        f.write(r.content)
    dem_array, geo_transform = load_dem_data(dem_file_path)
    csv_file='/content/drive/My Drive/kusatsu-trajectory_simulation_results_20240312_020152.csv' #草津白根　99m/s
    df_results = pd.read_csv(csv_file)
    vent_locations_list = [[2, -115687, 70449, 2045]]

BUN = 2
azimuths = np.linspace(0, 360, BUN, endpoint=False)

collision_details_df = plot_all_trajectories_and_terrain(df_results, dem_array, geo_transform, vent_locations_list[0][1], vent_locations_list[0][2], vent_locations_list[0][3], azimuths)

# Azimuthごとにグループ化し、各グループのデータを表示
for azimuth, group in collision_details_df.groupby('Azimuth'):
    print(f"Azimuth: {azimuth} degrees")
    display(group)  # Jupyter NotebookやGoogle Colabなどの環境で表を綺麗に表示
    print("\n")  # グループ間に空行を挿入

# 落下角度で色分けしてプロット
plot_shaded_dem(dem_array, geo_transform, collision_details_df, 'Fall Angle', 'Fall Angle Colored', cmap='jet_r')

# 落下速度で色分けしてプロット
plot_shaded_dem(dem_array, geo_transform, collision_details_df, 'Fall Speed', 'Fall Speed Colored', cmap='jet')
```
