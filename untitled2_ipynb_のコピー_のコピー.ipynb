```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

def calculate_trajectory(v0, angle_deg, size, density, Cd, ground_level):
    angle_rad = np.radians(angle_deg)
    v = v0 * np.sin(angle_rad)**1.5
    v0x = v * np.cos(angle_rad)
    v0y = v * np.sin(angle_rad)
    initial_conditions = [0, 0, v0x, v0y]

    def equations_of_motion(t, y):
        x, y, vx, vy = y
        v = np.sqrt(vx**2 + vy**2)
        Fd = 0.5 * Cd * Dens * (np.pi * (size / 2)**2) * v**2
        dvxdt = -Fd * vx / (v * density * 4/3 * np.pi * (size / 2)**3)
        dvydt = -9.81 - Fd * vy / (v * density * 4/3 * np.pi * (size / 2)**3)
        return [vx, vy, dvxdt, dvydt]

    def height_event(t, y):
        return y[1] - ground_level
    height_event.terminal = True
    height_event.direction = -1

    result = solve_ivp(equations_of_motion, [0, np.inf], initial_conditions,
                       method='RK45', events=height_event, max_step=0.1, dense_output=True)
    sol = result.sol(result.t_events[0]) if len(result.t_events[0]) > 0 else None
    collision_point = (sol[0], sol[1]) if sol is not None else (None, None)

    return result, collision_point

# 
P0 = 101000  #  (Pa)
deg=10.0
T0 = deg+273.15  #  (K)
L = 0.0065  #  (K/m)
g = 9.80665  #  (m/s^2)
M = 0.0289644  #  (kg/mol)
R = 8.3144598  #  (J/(molK))
h = 3500  #  (m)

# 3000m
T = T0 - L * h

# 3000m
P = P0 * (1 - L * h / T0) ** (g * M / (R * L))

# 3000m
Dens = P * M / (R * T)

size = 0.3
density = 2200
Cd = 0.47
V0dmin =188
d_max = 1.15
B = 1.5

ground_levels=[100]

plt.figure(figsize=(10, 6))
angles = np.arange(40, 81, 1)  # 1

for ground_levlel in ground_levels:

    max_distance = 0
    best_angle = None
    trajectories = {}

    for angle in angles:
        v0 = V0dmin * (1 - (size / d_max)**B)
        result, collision_point = calculate_trajectory(v0, angle, size, density, Cd,  ground_levlel )
        trajectories[angle] = {'x': result.y[0], 'y': result.y[1], 'collision_point': collision_point}
        distance = collision_point[0] if collision_point[0] is not None else 0
        if distance > max_distance:
            max_distance = distance
            best_angle = angle

    # 
    best_result, best_collision_point = calculate_trajectory(v0, best_angle, size, density, Cd,  ground_levlel )

    plt.plot(best_result.y[0], best_result.y[1], label=f'Best Angle: {best_angle}, Max Distance: {max_distance[0]:.2f} m', linewidth=2, color='red')

plt.title('Projectile Trajectories with Air Resistance and Optimal Angle')
plt.xlabel('Distance (m)')
plt.ylabel('Height (m)')
plt.legend()
plt.grid(True)
plt.gca().set_aspect('equal', adjustable='box')

plt.show()
```
