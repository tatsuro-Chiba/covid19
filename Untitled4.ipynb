```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from osgeo import gdal
import urllib.request
import math
from scipy.interpolate import interp1d


# calculate_fall_point関数の定義
def calculate_fall_point(vent_x, vent_y, bearing, distance):
    bearing_rad = math.radians(bearing)
    delta_x = distance * math.cos(bearing_rad)
    delta_y = distance * math.sin(bearing_rad)
    return vent_x + delta_x, vent_y + delta_y

# DEMファイルのダウンロードと読み込み
dem_url = "https://paspro.com/bousai/kazan/_/bbs/?file_id=MV8wMDA2MDQ%3D&name=hakone-5m-wide-3.tif"
#dem_file = "/mnt/f/program/ballistics/input/hakone-5m-wide-3.tif"
dem_file="dem.tif"
urllib.request.urlretrieve(dem_url, dem_file)
dem_data = gdal.Open(dem_file)
dem_band = dem_data.GetRasterBand(1) # Get the raster band
dem_array = dem_band.ReadAsArray()
geo_transform = dem_data.GetGeoTransform()


# 噴火口の位置情報
vent_locations_list = [
    [10.0, -74266.64, -83016.92, 916.5]]
'''
    [18.0, -74766.64, -83216.92, 941.0],
    [48.0, -73866.64, -83516.92, 958.7],
    [57.0, -74366.64, -83716.92, 994.6],
    [87.0, -73466.64, -84016.92, 1229.0],
    [99.0, -73766.64, -84316.92, 1263.9]
]
'''


# 陰影図の生成（簡易版）
def hillshade(array, azimuth, angle_altitude):
    x, y = np.gradient(array)
    slope = np.pi / 2. - np.arctan(np.sqrt(x*x + y*y))
    aspect = np.arctan2(-x, y)
    azimuth_rad = azimuth * np.pi / 180.
    angle_altitude_rad = angle_altitude * np.pi / 180.

    shaded = np.sin(angle_altitude_rad) * np.sin(slope) \
             + np.cos(angle_altitude_rad) * np.cos(slope) \
             * np.cos(azimuth_rad - aspect)
    return 255 * (shaded + 1) / 2

hillshade_array = hillshade(dem_array, 315, 45)  # 315°で光源、高度角45°


# 色のリストを定義（例: 6つの火口に対して6色を定義）
colors = ['red', 'blue', 'green', 'yellow', 'purple', 'orange']

# Get spatial reference information
geo_transform = dem_data.GetGeoTransform()
# 陰影図の表示
plt.figure(figsize=(8, 8))
extent = [
    geo_transform[0],
    geo_transform[0] + dem_array.shape[1] * geo_transform[1],  # X軸の最大値
    geo_transform[3] + dem_array.shape[0] * geo_transform[5],  # Y軸の最小値
    geo_transform[3]  # Y軸の最大値
]
plt.imshow(hillshade_array, cmap='gray', extent=extent)
plt.colorbar(label='Intensity')

# 噴火口を地理座標系でプロット
for i, vent in enumerate(vent_locations_list):
    vent_id, vent_x, vent_y, _ = vent
    # 地理座標をそのまま使用
    plt.plot(vent_x, vent_y, marker='o', color=colors[i % len(colors)], linestyle='None', markersize=5, label=f'Vent ID {vent_id}')

# 凡例を表示
plt.legend()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
#plt.show()

# 軌道データのダウンロードと読み込み

trajec_url = "https://paspro.com/bousai/kazan/_/bbs/?file_id=MV8wMDA3MTA%3D&name=simulation_v0z%3D210.00_size%3D0.3_density%3D2200_CD%3D0.47_20240214_210641.csv"
#trajec_url = "https://paspro.com/bousai/kazan/_/bbs/?file_id=MV8wMDA3MjE%3D&name=extended_resampled_trajectories+_v210.csv"
#trajec_url = "https://paspro.com/bousai/kazan/_/bbs/?file_id=MV8wMDA3MjI%3D&name=extended_resampled_trajectories_v100.csv"
trajec_file="trajec_file.csv"
#trajec_file = "/mnt/f/program/ballistics/extended_resampled_trajectories_v100.csv"
urllib.request.urlretrieve(trajec_url, trajec_file)

data = pd.read_csv(trajec_file)

# 内挿補間されたデータを保持するリストを初期化
interpolated_data = []

# 各発射角度でデータを処理
for angle in data['Launch Angle (deg)'].unique():
    # 現在の角度のデータをフィルタリング
    angle_data = data[data['Launch Angle (deg)'] == angle]

    # 内挿準備
    if len(angle_data['X Position (m)']) > 1:  # 少なくとも2点必要
        x_positions = angle_data['X Position (m)']

        # 各パラメーターの内挿関数を定義
        z_interp = interp1d(x_positions, angle_data['Z Position (m)'], fill_value='extrapolate')
        x_vel_interp = interp1d(x_positions, angle_data['X Velocity (m/s)'], fill_value='extrapolate')
        z_vel_interp = interp1d(x_positions, angle_data['Z Velocity (m/s)'], fill_value='extrapolate')
        speed_interp = interp1d(x_positions, angle_data['Speed (m/s)'], fill_value='extrapolate')
        flight_angle_interp = interp1d(x_positions, np.degrees(np.arctan2(angle_data['Z Velocity (m/s)'], angle_data['X Velocity (m/s)'])), fill_value='extrapolate')

        # 1m間隔で新しいX位置を生成
        new_x_positions = np.arange(np.ceil(x_positions.min()), np.floor(x_positions.max()) + 1, 1)

        # 各パラメーターの内挿を実行
        new_z_positions = z_interp(new_x_positions)
        new_x_velocities = x_vel_interp(new_x_positions)
        new_z_velocities = z_vel_interp(new_x_positions)
        new_speeds = speed_interp(new_x_positions)
        new_flight_angles = flight_angle_interp(new_x_positions)

        # 現在の角度の内挿データをリストに追加
        for x, z, xv, zv, s, fa in zip(new_x_positions, new_z_positions, new_x_velocities, new_z_velocities, new_speeds, new_flight_angles):
            interpolated_data.append([angle, x, z, xv, zv, s, fa])

# リストをDataFrameに変換し、数値を4桁に丸める
interpolated_trajec_data = pd.DataFrame(interpolated_data, columns=['Launch Angle (deg)', 'X Position (m)', 'Z Position (m)', 'X Velocity (m/s)', 'Z Velocity (m/s)', 'Speed (m/s)', 'Flight Angle (deg)']).round(4)

# 最初の5行を表示
#print(interpolated_trajec_data.head())

# ここでinterpolated_dfを使用
# 地形データのダウンロードと読み込み
profile_url = "https://paspro.com/bousai/kazan/_/bbs/?file_id=MV8wMDA3MjM%3D&name=xyz_terrain_profiles_20240218_022752.csv"
#profile_file = "/mnt/f/program/ballistics/tikei/xyz_terrain_profiles_all_360.csv"
profile_file="profile_file.csv"
urllib.request.urlretrieve(profile_url, profile_file)
terrain_data = pd.read_csv(profile_file)

terrain_profile = terrain_data.iloc[1, 5:].astype(float)

# 断面図の左の最初の点の高さをbase_heightとして設定
base_height = terrain_profile.iloc[0]

# 最大水平距離のディクショナリを準備
max_horizontal_distances = {}
launch_angles = trajec_data['Launch Angle (deg)'].unique()

# 落下地点の情報を格納するための空のリストを用意
fall_points_info = []
max_distance_info = []
# 最大到達距離とその時の情報を格納するリストを初期化
max_distance_info_per_section = []

BUN = 360 #地形断面の分割数
num = BUN*102 #分割数×火口数=解析断面数
num=24

for i in range(num):  # 複数断面に対する解析
    # 地形プロファイルの読み込み
    terrain_profile = terrain_data.iloc[i, 5:].astype(float)
    AAzimth = terrain_data.iloc[i, 4]  # 5番目のカラムの値を取得
    vid = terrain_data.iloc[i,0]  # 1番目のカラムの値を取得
    # base_heightの設定: 地形プロファイルの最初の点の高さを使用
    base_height = terrain_profile.iloc[0]
    # 各発射角度に対する分析
    # 各発射角度での最大水平距離を記録する辞書をリセット
    max_horizontal_distances = {}

    collision_flag = False  # 衝突検出フラグをクリア
    # 最大到達距離とその時の情報を格納するリストを初期化
    max_distance_info_per_section = []

    for angle in launch_angles:
        if 40 <= angle <= 80:

            trajectory_subset = trajec_data[trajec_data['Launch Angle (deg)'] == angle]
            adjusted_trajectory_height = trajectory_subset['Z Position (m)'] + base_height

            terrain_profile_aligned = terrain_profile[:len(adjusted_trajectory_height)].reset_index(drop=True)
            collision_indices = (adjusted_trajectory_height.values < terrain_profile_aligned.values)

            if any(collision_indices):
                first_collision_index = np.argmax(collision_indices)
                collision_flag = True

                # 衝突時点の情報を取得
                vx_collision = trajectory_subset.iloc[first_collision_index]['X Velocity (m/s)']
                vz_collision = trajectory_subset.iloc[first_collision_index]['Z Velocity (m/s)']
                v_total = math.sqrt(vx_collision**2 + vz_collision**2)
                theta_collision = math.degrees(math.atan2(vz_collision, vx_collision))
                max_distance = trajectory_subset.iloc[first_collision_index]['X Position (m)']

                # この情報を辞書に格納
                max_horizontal_distances[angle] = (max_distance, v_total, theta_collision)


                # 落下角度を計算（角度は度数法で）
                theta_collision = math.degrees(math.atan2(vz_collision, vx_collision))

                fall_point_x, fall_point_y = calculate_fall_point(vent_x, vent_y, AAzimth , trajectory_subset.iloc[first_collision_index]['X Position (m)'])

                # 衝突点の情報をリストに追加
                fall_points_info.append([
                    fall_point_x, fall_point_y, base_height,  # 地理的なXYと高さ
                    vent_id, AAzimth, angle, trajectory_subset.iloc[first_collision_index]['X Position (m)'],  # 火口ID、方位、射出角、距離
                    v_total, theta_collision  # 落下時の速度と角度
                ])

                 # 衝突判定があれば、その情報をmax_horizontal_distancesに格納
                max_horizontal_distances[angle] = (max_distance, v_total, theta_collision)

    # 各断面での最大到達距離を出した角度を認定
    if max_horizontal_distances:
        max_distance_angle = max(max_horizontal_distances, key=max_horizontal_distances.get)
        max_distance, v_total, theta_collision = max_horizontal_distances[max_distance_angle]




        # この断面での最大到達距離とその時の情報を格納
        max_distance_info_per_section.append({
            'section': i,
            'angle': max_distance_angle,
            'max_distance': max_distance,
            'v_total': v_total,
            'theta_collision': theta_collision
        })

        # グラフのタイトルに含める情報を取得
        angle = max_distance_angle
        max_distance = max_distance
        v_total = v_total
        theta_collision = theta_collision
    else:
        # max_horizontal_distances が空の場合の処理（例: ログ出力、デフォルト値の使用など）
        print(f"断面 {i+1} で有効な衝突が検出されませんでした。")



    # 最大水平距離を持つ発射角度の検出
    max_distance_angle = max(max_horizontal_distances, key=max_horizontal_distances.get)
    max_distance = max_horizontal_distances[max_distance_angle]

    # 第2パラメータ（方位）を使用して最終落下地点を計算
    profile_row = terrain_data.iloc[i]  # 仮に2行目を使用
    vent_x, vent_y, bearing = profile_row[1], profile_row[2], profile_row[4]  # X, Y座標と方位

    # 最適な軌道の最大水平距離を使用して最終落下地点を計算（ここでは簡単のためmax_distanceとbearingを使用）
    max_distance = 1000  # 仮の最大水平距離
    fall_point_x, fall_point_y = calculate_fall_point(vent_x, vent_y, bearing, max_distance)

    # 最適な軌道を持つ発射角度とその最大到達距離#######################################################################
    max_distance_angle = max(max_horizontal_distances, key=max_horizontal_distances.get)
    max_distance = max_horizontal_distances[max_distance_angle]

    # ここで最大距離のデータをmax_distance_infoリストに追加
    max_distance_info.append([
        fall_point_x, fall_point_y, base_height,  # 地理的なXYと高さ
        vent_id, bearing, angle, trajectory_subset.iloc[first_collision_index]['X Position (m)'],  # 火口ID、方位、射出角、距離
        v_total, theta_collision  # 落下時の速度と角度
    ])


    # 火口IDと方位を取得（ここでは例として最初の火口を使用）
    vent_id = vent_locations_list[0][0]  # 火口ID
    bearing = profile_row[4]  # 方位は地形プロファイルデータから取得


    # グラフの描画
    plt.figure(figsize=(8, 5))

    terrain_x = np.linspace(0, len(terrain_profile) - 1, len(terrain_profile))
    plt.plot(terrain_x, terrain_profile, label='Terrain Profile', color='black', linewidth=2)

    # 最適な軌道データの取得
    trajectory_subset_max = trajec_data[trajec_data['Launch Angle (deg)'] == max_distance_angle]
    adjusted_height_max = trajectory_subset_max['Z Position (m)'] + base_height
    plt.plot(trajectory_subset_max['X Position (m)'], adjusted_height_max, label=f'Angle {max_distance_angle:.0f}°', color='blue')

    # 発射点の表示
    plt.scatter(trajectory_subset_max['X Position (m)'].iloc[0], adjusted_height_max.iloc[0], color='red', s=60, marker='o', label='Launch Point')





    # 衝突地点の最適な発射角度での表示
    if max_distance_angle in max_horizontal_distances:

        # 最大水平距離を持つ発射角度の検出
        # Correctly extract the maximum distance value before passing it to calculate_fall_point
        # max_distanceがタプルである場合、最初の要素（距離）を取り出します
        # max_distance_value = max_horizontal_distances[max_distance_angle][0]
        # Before calling the function, print out the value to verify it
        # print(f"max_distance before function call: {max_distance}")
        # Extract just the distance value from the tuple

        # Correctly extract the maximum distance value before passing it to calculate_fall_point
        max_distance_value = max_horizontal_distances[max_distance_angle][0]
        fall_point_x, fall_point_y = calculate_fall_point(vent_x, vent_y, bearing, max_distance_value)


        # This is the corrected line where you get the maximum horizontal distance
        collision_x_position = max_horizontal_distances[max_distance_angle][0]  # Get only the distance

        # Now you can safely convert it to an integer if necessary
        collision_x_position = int(collision_x_position)
        collision_y_position = terrain_profile[collision_x_position]
        plt.scatter(collision_x_position, collision_y_position, color='green', s=60, marker='o', label='Collision Point')

    # 例: 第1の断面に対するグラフを描画する場合
    section_index = 0  # 0ベースのインデックスを使用 (1つ目の断面に対応)

    # max_distance_info_per_section から対象の断面情報を取得
    info = max_distance_info_per_section[section_index]

    # タイトルに含める情報を取得
    angle = info['angle']
    max_distance = info['max_distance']
    v_total = info['v_total']
    theta_collision = info['theta_collision']

    # グラフのタイトルを設定
    #title = f"断面 {section_index + 1}: 角度 {angle}度, 最大到達距離 {max_distance}m, 速度 {v_total}m/s, 落下角度 {theta_collision}度"

    # グラフのタイトルにこれらの情報を含める
    title = f"Hakone Vent ID ={vent_id:.0f} : {AAzimth:.0f} deg L ={max_distance:.0f} m H ={collision_y_position:.0f} m {v_total:.0f} m/s {theta_collision:.0f} deg"

    # グラフの描画
    # ... ここにグラフ描画のためのコード ...
    plt.title(title)
    plt.xlabel('X Position (m)')
    plt.ylabel('Z Position (m)')
    plt.legend()
    plt.grid(True)

    #plt.savefig(f"plot_hakone-id_{vid}_{AAzimth}_v0_210.png")
    plt.show()
    # plt.close()
    profile_row = terrain_data.iloc[i]  # 現在の断面データ
    vent_id = profile_row['Vent ID']
    bearing = AAzimth
    launch_angle = max_distance_angle  # これは前のコードからの最適な発射角度

    #velocity_at_impact = 0  # 実際の速度データに置き換える

    #angle_at_impact = 0  # 実際の角度データに置き換える


# 最大距離のデータフレームを作成（以前のコードから）
max_distance_df = pd.DataFrame(max_distance_info, columns=[
    'Fall_Point_X', 'Fall_Point_Y', 'Height',
    'Vent_ID', 'Bearing', 'Launch_Angle', 'Max_Distance_from_Vent',
    'Velocity_at_Impact', 'Angle_at_Impact'
])

# すべての衝突点のデータフレームを作成（新しく追加したコード）
all_collisions_df = pd.DataFrame(fall_points_info, columns=[
    'Fall_Point_X', 'Fall_Point_Y', 'Height',
    'Vent_ID', 'Bearing', 'Launch_Angle', 'Distance_from_Vent',
    'Velocity_at_Impact', 'Angle_at_Impact'
])

# ファイル名の定義
max_distance_filename = "fall_points_max_distance.csv"
all_collisions_filename = "fall_points_all_collisions.csv"

# CSVファイルとしてエクスポート
max_distance_df.to_csv(max_distance_filename, index=False)
all_collisions_df.to_csv(all_collisions_filename, index=False)



# データフレームの例。実際のデータに置き換えてください。
fall_points_df = pd.DataFrame(all_collisions_df)

# 落下速度の最小値と最大値を取得
vmin = fall_points_df['Velocity_at_Impact'].min()
vmax = fall_points_df['Velocity_at_Impact'].max()

# カラーマップを作成
cmap = plt.cm.viridis

# プロットの作成
plt.figure(figsize=(10, 6))

extent = [
    geo_transform[0],
    geo_transform[0] + dem_array.shape[1] * geo_transform[1],  # X軸の最大値
    geo_transform[3] + dem_array.shape[0] * geo_transform[5],  # Y軸の最小値
    geo_transform[3]  # Y軸の最大値
]
plt.imshow(hillshade_array, cmap='gray', extent=extent)

# 噴火口を地理座標系でプロット
for i, vent in enumerate(vent_locations_list):
    vent_id, vent_x, vent_y, _ = vent
    # 地理座標をそのまま使用
    plt.plot(vent_x, vent_y, marker='o', color=colors[i % len(colors)], linestyle='None', markersize=5, label=f'Vent ID {vent_id}')

# 各落下地点をプロット（マーカーのサイズを小さくする）
sc = plt.scatter(fall_points_df['Fall_Point_X'], fall_points_df['Fall_Point_Y'],
                 c=fall_points_df['Velocity_at_Impact'], cmap=cmap, vmin=vmin, vmax=vmax, s=10)  # s=10でマーカーサイズを小さく設定

# カラーバーを追加
cbar = plt.colorbar(sc)
cbar.set_label('Velocity at Impact (m/s)')
title = f"Fall Points with Velocity  ID:{vent_id:.0f}   V0maxVt:210 m/s   size:30 cm   Cd:0.47"

plt.title(title)
plt.xlabel('X Position (m)')
plt.ylabel('Y Position (m)')
plt.grid(True)
plt.show()

```
